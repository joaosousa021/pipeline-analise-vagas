EXECUTE ABAIXO PARA VISUALIZAR UMA TABELA DE VAGAS (SHIFT + ENTER)
--------------------------------------------------

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect('vagas.db')


df = pd.read_sql('SELECT * FROM vagas', conn)
conn.close()

print("Dados carregados do banco!")
df.head()
```

----------------------------------------------
EXECUTE ABAIXO PARA UMA CONTAGEM DE SKILLS (SHIFT + ENTER)
-

```python
colunas_skills = [col for col in df.columns if col.startswith('tem_')]
skills_soma = df[colunas_skills].sum().sort_values(ascending=False)

print("Contagem de cada skill:")
print(skills_soma)
```

----------------------------------------------
EXECUTE ABAIXO PARA UMA CONTAGEM DE SKILLS (SHIFT + ENTER)
-

```python
plt.style.use('ggplot')
skills_soma.plot(kind='bar', figsize=(12, 7))

plt.title('Top Habilidades Pedidas em Vagas de Dados', fontsize=16)
plt.ylabel('Número de Vagas', fontsize=12)
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 

plt.savefig('top_skills.png')

plt.show()
```
